# Cluster map

```python
import pandas as pd
from techminer import RecordsDataFrame

rdf = RecordsDataFrame(
    pd.read_json(
        'step-07.json', 
        orient='records', 
        lines=True))
```

```python
rdf.autocorr(
    column='Authors',
    sep=',',
    top_n=30
).map(
    figsize=(11,11)
)
```
