```python
import pandas as pd
from techminer import RecordsDataFrame

rdf = RecordsDataFrame(
    pd.read_json(
        'step-07.json', 
        orient='records', 
        lines=True))
```

```python
id = rdf.terms_by_year( 
    column='Source title', 
    sep=None,
    top_n=10)['ID'].explode().tolist()
```
