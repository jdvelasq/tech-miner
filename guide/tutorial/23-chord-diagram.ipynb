# Chord diagram

```python
import pandas as pd
from techminer import RecordsDataFrame

rdf = RecordsDataFrame(
    pd.read_json(
        'step-07.json', 
        orient='records', 
        lines=True))
```

```python
rdf.autocorr(
    column='Authors',
    sep=',',
    top_n=20).chord_diagram(minval=0.50)
```

```python
rdf.autocorr(
    column='Authors',
    sep=',',
    top_n=20
).map_to_chord()
```
