# Cluster map - Co-ocurrences

```python
import pandas as pd
from techminer import RecordsDataFrame

rdf = RecordsDataFrame(
    pd.read_json(
        'step-07.json', 
        orient='records', 
        lines=True))
```

```python
rdf.co_ocurrence(
    column_r='Authors', 
    column_c='Authors', 
    sep_r=',', 
    sep_c=',',
    top_n=10
).heatmap()
```

```python
rdf.co_ocurrence(
    column_r='Authors', 
    column_c='Authors', 
    sep_r=',', 
    sep_c=',',
    top_n=10
).cluster_map(
    figsize=(11,11),
    font_size=10,
    factor = 0.1,
    size=(300,1000)
)
```

```python
rdf.co_ocurrence(
    column_r='Authors', 
    column_c='keywords (cleaned)', 
    sep_r=',', 
    sep_c=';',
    top_n=10
).heatmap()
```

```python
rdf.co_ocurrence(
    column_r='Source title', 
    column_c='keywords (cleaned)', 
    sep_r=None, 
    sep_c=';',
    top_n=3
).cluster_map(
    figsize=(11,11),
    font_size=10,
    factor = 0.1,
    size=(200,800)
)
```
