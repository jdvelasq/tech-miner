# Cluster map - Co-ocurrences

```python
import pandas as pd
from techminer import RecordsDataFrame

rdf = RecordsDataFrame(
    pd.read_json(
        'step-07.json', 
        orient='records', 
        lines=True))
```

```python
rdf.terms_by_terms(
    column_r='Source title', 
    column_c='keywords (cleaned)', 
    sep_r=None, 
    sep_c=';',
    top_n=10
).cluster_map(
    figsize=(11,11),
    font_size=10,
    factor = 0.1,
    size=(30,800)
)
```
