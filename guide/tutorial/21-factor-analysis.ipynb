# Crosscorrelation matrix

```python
import pandas as pd
from techminer import RecordsDataFrame

rdf = RecordsDataFrame(
    pd.read_json(
        'step-07.json', 
        orient='records', 
        lines=True))
```

```python
rdf.factor_analysis(
    column='Authors', 
    sep=',', 
    n_components=5, 
    top_n=40).tomatrix()
```
