# Second level analysis

In this part, a second level analysis of the bibliographical information is done.

```python
import pandas as pd
from techminer import RecordsDataFrame
records = pd.read_json('cleaned-records.json', orient='records', lines=True)
rdf = RecordsDataFrame(records)
```

## Keywords per year

```python
rdf.term_by_term(
    termA='keywords (cleaned)', 
    termB='Year', 
    sepA=';', 
    sepB=None).head()
```

```python
rdf.term_by_term( 
    termA='keywords (cleaned)', 
    termB='Year', 
    sepA=';', 
    sepB=None, 
    minmax=(2,3)).head()   ## minimum and maximum values in the matrix
```

```python
rdf.term_by_term( 
    termA='keywords (cleaned)', 
    termB='Year', 
    sepA=';', 
    sepB=None, 
    minmax=(5,8)).heatmap(ascendingA=True, ascendingB=False)
```

```python
rdf.term_by_term( 
    termA='keywords (cleaned)', 
    termB='Year', 
    sepA=';', 
    sepB=None, 
    minmax=(5,8)).heatmap_in_altair(ascendingA=True, ascendingB=False)
```

```python
rdf.term_by_term( 
    termA='keywords (cleaned)', 
    termB='Year', 
    sepA=';', 
    sepB=None, 
    minmax=(5,8)).circleplot_in_altair(ascendingA=True, ascendingB=False)
```

```python
rdf.term_by_term( 
    termA='keywords (cleaned)', 
    termB='Year', 
    sepA=';', 
    sepB=None, 
    minmax=(5,8)).to_matrix()
```

```python
centers, clusters = rdf.term_by_term( 
    termA='Authors', 
    termB='keywords (cleaned)', 
    sepA=',', 
    sepB=';').kmeans(n_clusters = 3)
```

```python
centers
```

```python
centers.transpose().heatmap()
```

```python
clusters.sort_values('cluster')
```

## Autocorrelation

```python
rdf.autocorrelation(
    term='keywords (cleaned)',
    sep=';'
).to_matrix()
```

```python
rdf.autocorrelation(
    term='keywords (cleaned)',
    sep=';'
).circleplot_in_altair()
```

```python
rdf.autocorrelation(
    term='keywords (cleaned)',
    sep=';').heatmap()
```

## Crosscorrelation

```python
rdf.crosscorrelation(
    termA='keywords (cleaned)',
    sepA=';',
    termB='Authors',
    sepB=','
).head(10)
```

```python
rdf.crosscorrelation(
    termA='keywords (cleaned)',
    sepA=';',
    termB='Authors',
    sepB=','
).heatmap()
```

```python
rdf.crosscorrelation(
    termA='keywords (cleaned)',
    sepA=';',
    termB='Authors',
    sepB=','
).to_matrix()
```

```python
centers, clusters = rdf.crosscorrelation(
    termA='keywords (cleaned)',
    sepA=';',
    termB='Authors',
    sepB=','
).to_matrix().kmeans(n_clusters = 3)
```

```python
centers
```

```python
centers.transpose().heatmap()
```

## Factor analysis

```python
rdf.factor(
    term='keywords (cleaned)', 
    sep=';', 
    n_components=5, 
    N=40).transpose().heatmap()
```
