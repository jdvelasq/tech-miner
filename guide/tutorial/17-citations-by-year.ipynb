# Citations by year

```python
import pandas as pd
from techminer import RecordsDataFrame

rdf = RecordsDataFrame(
    pd.read_json(
        'step-07.json', 
        orient='records', 
        lines=True))
```

```python
rdf.citations_by_year()
```

```python
rdf.citations_by_year().barhplot()
```

```python
rdf.citations_by_year().barhplot('altair')
```

```python
rdf.citations_by_year(cumulative=True).barhplot('altair')
```
