# Sankey plot

```python
import pandas as pd
from techminer import RecordsDataFrame

rdf = RecordsDataFrame(
    pd.read_json(
        'step-07.json', 
        orient='records', 
        lines=True))
```

```python
rdf.cross_corr(
    column_r='keywords (cleaned)',
    sep_r=';',
    column_c='Authors',
    sep_c=','
).sankey_plot(minval=0.1)
```
