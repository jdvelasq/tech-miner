# Terms by terms by year

```python
import pandas as pd
from techminer import RecordsDataFrame

rdf = RecordsDataFrame(
    pd.read_json(
        'step-07.json', 
        orient='records', 
        lines=True))
```

```python
rdf.terms_by_terms_by_year(
    column_r='Authors', 
    column_c='keywords (cleaned)', 
    sep_r=',', 
    sep_c=';',
    top_n=5)
```

```python
rdf.terms_by_terms_by_year(
    column_r='Authors', 
    column_c='keywords (cleaned)', 
    sep_r=',', 
    sep_c=';',
    top_n=5).print_IDs()
```
