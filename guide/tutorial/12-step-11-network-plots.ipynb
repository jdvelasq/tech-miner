```python
import pandas as pd
from techminer import RecordsDataFrame

rdf = RecordsDataFrame(
    pd.read_json(
        'step-06.json', 
        orient='records', 
        lines=True))
```

```python
rdf.term_by_term(
    termA='Authors', 
    termB='Author Keywords', 
    sepA=',', 
    sepB=';',
    minmax=(2,4)).to_matrix()
```
