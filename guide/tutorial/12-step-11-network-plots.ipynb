# Step 11: Map plots

```python
import pandas as pd
from techminer import RecordsDataFrame

rdf = RecordsDataFrame(
    pd.read_json(
        'step-06.json', 
        orient='records', 
        lines=True))
```

```python
rdf.autocorrelation(
    term='Authors',
    sep=',',
    N=50).map()
```

```python
rdf.crosscorrelation(
    termA='Authors',
    termB='keywords (cleaned)',
    sepA=',',
    sepB=';',
    N=80).map()
```

```python
rdf.factor_analysis(
    term='keywords (cleaned)',
    sep=';',
    n_components = 10,
    N=30).map()
```
