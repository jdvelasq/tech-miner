# Heatmap

```python
import pandas as pd
from techminer import RecordsDataFrame

rdf = RecordsDataFrame(
    pd.read_json(
        'step-07.json', 
        orient='records', 
        lines=True))
```

```python
rdf.terms_by_year( 
    column='Authors', 
    sep=',',
    top_n=20).heatmap(figsize=(8,2))
```
