# Crosscorrelation matrix

```python
import pandas as pd
from techminer import RecordsDataFrame

rdf = RecordsDataFrame(
    pd.read_json(
        'step-07.json', 
        orient='records', 
        lines=True))
```

```python
rdf.cross_corr(
    column_r='keywords (cleaned)',
    sep_r=';',
    column_c='Authors',
    sep_c=',',
    top_n=20
).head(30)
```

```python
rdf.documents_by_terms('Authors', sep=',', top_n=20)
```

```python
rdf.cross_corr(
    column_r='keywords (cleaned)',
    sep_r=';',
    column_c='Authors',
    sep_c=',',
    top_n=20
).heatmap()
```

```python
rdf.cross_corr(
    column_r='keywords (cleaned)',
    sep_r=';',
    column_c='Authors',
    sep_c=',',
    top_n=20
).print_IDs()
```
