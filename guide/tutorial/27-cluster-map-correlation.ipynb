# Cluster map - correlations

```python
import pandas as pd
from techminer import RecordsDataFrame

rdf = RecordsDataFrame(
    pd.read_json(
        'step-07.json', 
        orient='records', 
        lines=True))
```

```python
rdf.autocorr(
    column='Authors',
    sep=',',
    top_n=30
).cluster_map(
    figsize=(11,11),
    font_size=10,
    factor = 0.1,
    size=(30,800)
)
```
