# Terms by terms

```python
import pandas as pd
from techminer import RecordsDataFrame

rdf = RecordsDataFrame(
    pd.read_json(
        'step-07.json', 
        orient='records', 
        lines=True))
```

```python
rdf.terms_by_terms(
    column_r='Source title', 
    column_c='keywords (cleaned)', 
    sep_r=None, 
    sep_c=';',
    top_n=20).head()
```

```python
rdf.terms_by_terms(
    column_r='Source title', 
    column_c='keywords (cleaned)', 
    sep_r=None, 
    sep_c=';',
    top_n=10).tomatrix()
```
