```python
import pandas as pd
from techminer import RecordsDataFrame

rdf = RecordsDataFrame(
    pd.read_json(
        'step-07.json', 
        orient='records', 
        lines=True))
```

```python
rdf.documents_by_terms('Authors', sep=',').head(16)
```

```python
rdf.documents_by_terms('Authors', sep=',').print_IDs()
```
